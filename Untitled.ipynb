```python
import pandas as pd

df = pd.read_excel('database organised.xlsx')
df['Name'] = df['Name'].fillna("(no name available)")
df['Address'] = df['Address'].fillna("(no address available)")
df['Postcode'] = df['Postcode'].fillna("(none)")
df = df[df['Amount'].notnull()]
df['Reason'] = df['Reason'].fillna("(no source available)")
df = df[[u'Name', u'Address', u'Postcode', u'Amount', u'Reason']]
df.head()
```

```python
df.sort_values('Amount',ascending=True)
```

```python
df.columns
```

```python
df.to_json('db.json',orient='values')
```

```python
lis = ''
for row in df.values:
    lis = lis + "<li>$"+unicode(int(row[4]))+' - '+row[1].strip()+"</li>"
```
